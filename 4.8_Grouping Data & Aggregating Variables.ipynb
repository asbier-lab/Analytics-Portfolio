```python
# Import 
import pandas as pd
import numpy as np

import os

#  load the pickle file
path = '/Users/am/Desktop/CF Data Analysis essentials /2_Data Analytics Immersions_2 von 3/4 Phyton/4.7/Prepared Data'
file_path = os.path.join(path, 'ords_prods_merge_updated.pkl')

# Load the dataframe
df = pd.read_pickle(file_path)

# Check the shape and a preview
print(df.shape)
df.head()

print(df.columns.tolist())
```

```python
df.groupby('product_name')
```

```python
df['mean_product_price'] = df.groupby(['user_id'])['prices'].transform('mean')
df.loc[df['mean_product_price']<10,'spending_flag']='Low_spender'
df.loc[df['mean_product_price']>= 10, 'spending_flag']='High_spender'

df['spending_flag'].value_counts(dropna = False)
```

```python
# Aggregating the mean order_number per department
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
df.groupby('department_id')['order_number'].mean()
```

```python
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)

df.head(17)
```

```python
df.head(100)
```

```python
pd.options.display.max_rows = None
```

```python
# Create the loyalty_flag column based on max_order
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check value counts for the new loyalty_flag column
print(df['loyalty_flag'].value_counts())
print(df.columns)
```

```python
# Calculate median days_since_prior_order per user
median_days = df.groupby('user_id')['days_since_prior_order'].transform('median')

# Create the frequency flag based on median_days
df.loc[median_days > 20, 'order_frequency_flag'] = 'Non-frequent customer'
df.loc[(median_days > 10) & (median_days <= 20), 'order_frequency_flag'] = 'Regular customer'
df.loc[median_days <= 10, 'order_frequency_flag'] = 'Frequent customer'

# Check value counts for the new loyalty_flag column
print(df['order_frequency_flag'].value_counts())
```
